```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
credits = pd.read_csv("/content/tmdb_5000_credits.csv", engine='python', on_bad_lines='warn')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on = 'title')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
def convert(text):
  l = []
  for i in  ast.literal_eval(text):
    l.append(i['name'])
  return l
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
def convert3(text):
  l = []
  counter = 0
  # Check if the text is already a list (meaning it was likely processed before)
  if isinstance(text, list):
    for item in text:
      if counter < 3:
        # Assuming items in the list are already the names if it's a list
        l.append(item)
        counter += 1
  elif isinstance(text, str):
    # Process as a string representation of a list of dictionaries
    list_of_dicts = ast.literal_eval(text)
    for i in list_of_dicts:
      if counter < 3:
        l.append(i['name'])
        counter += 1
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
  l = []
  if isinstance(text, str):
    list_of_dicts = ast.literal_eval(text)
    for i in list_of_dicts:
      if i['job'] == 'director':
        l.append(i['name'])
  elif isinstance(text, list):
    # If it's already a list, it means it's already been processed
    # and contains director names or is an empty list.
    # So, we return it as is, as it's the result of the desired operation.
    return text
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
from os import replace
def collapse(l):
  l1 = []
  for i in l:
    l1.append(i.replace(" ",""))
  return l1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies[['movie_id', 'title', 'tags']].copy()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features =5000,stop_words = 'english' )
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommended(movies):
  index = new[new['title']==movies].index[0]
  distances = similarity[index]
  movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(new.iloc[i[0]].title)
```

```python
recommended('Avatar')
```
